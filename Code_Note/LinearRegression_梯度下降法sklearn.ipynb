# 简单线性回归（梯度下降法）sklearn

### 0. 引入依赖

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. 导入数据 (data.csv)

```python
points = np.genfromtxt("data.csv",delimiter=",")

#points

#提取points中的两列数据分别作为x和y
x=points[:,0]
y=points[:,1]


plt.scatter(x,y)
plt.show()
```

### 2.定义损失函数

```python
#损失函数是系数的函数，另外还要传入数据x，y
def compute_cost(w,b,points):
    total_cost = 0
    M = len(points)

    #逐点计算平方误差，然后求平均数

    for i in range (M):
        x = points[i,0]
        y = points[i,1]
        total_cost+=(y-(w*x+b))**2
    
    return total_cost/M
```

#### 调库


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

x_new = x.reshape(-1,1) ##行数不限（-1），列数为一列， 把x变成二维的n*1矩阵
y_new = y.reshape(-1,1) ##行数不限（-1），列数为一列， 把x变成二维的n*1矩阵

lr.fit(x_new,y_new)
```

#### 从训练好的模型提取系数和截距

```python
w = lr.coef_[0][0]
b = lr.intercept_[0]

print("w is ", w)
print("b is ", b)

cost=compute_cost(w,b,points)

print("cost is ", cost) ##和最小二乘更为接近
```

### 3. 画出拟合曲线

```python
plt.scatter(x,y)

#针对每一个x，计算出预测的y值
pred_y=w*x+b

plt.plot(x,pred_y,c='r')
plt.show()
```
